```python
import pandas as pd
```

```python
%ls ../data/nyc/glossaries
```

```python
blobs = pd.read_json("../data/nyc/glossaries/blob.json")
```

```python
blobs.sample()
```

```python
links = pd.read_json("../data/nyc/glossaries/link.json")
```

```python
tables = pd.read_json("../data/nyc/glossaries/table.json")
```

```python
geospatials = pd.read_json("../data/nyc/glossaries/geospatial dataset.json")
```

```python
res = pd.concat([tables, geospatials, blobs, links])
```

```python
res = res.drop('flags', axis='columns')
```

```python
res = res.reset_index()
```
